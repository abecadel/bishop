Function we are looking for

\begin{align}
\sin 2 \pi x
\end{align}

```python
x_arr = np.linspace(0, 1, 100)
y_arr = np.array([math.sin(2*math.pi*y) for y in x_arr])
plt.plot(x_arr, y_arr, label="Without noise")

y_with_noise = y_arr + np.random.normal(0, 0.1, 100)
plt.plot(x_arr, y_with_noise, "ro", label="With noise")

plt.legend()
```

```python
def poly_fun(x, w_wektor):
    ret = 0
    
    for i, w in enumerate(w_wektor):
        ret += w * (x ** i)
    
    return ret

def error_fun(w, xn, tn):
    ret = 0
    
    for i, x in enumerate(xn):
        expected = tn[i]
        actuall = poly_fun(x, w)
        ret += (actuall - expected) ** 2
    
    return ret / 2

def root_mean_square_error_fun(w, xn, tn):
    return math.sqrt(2 * error_fun(w, xn, tn) / len(xn))

def show_polynomial(w, x_arr, y_arr):
    guess_y_arr = [poly_fun(x, w) for x in x_arr]
    root_error_y_arr = root_mean_square_error_fun(w, guess_y_arr, y_arr)
    plt.plot(x_arr,guess_y_arr, label=root_error_y_arr)
```

```python
w_M0 = [0.19]
show_polynomial(w_M0, x_arr, y_with_noise)

w_M1 = [0.82, -1.27]
show_polynomial(w_M1, x_arr, y_with_noise)

w_M3 = [0.31, 7.99, -25.43, 17.37]
show_polynomial(w_M3, x_arr, y_with_noise)

plt.plot(x_arr,y_with_noise, 'ro')

plt.legend()
```
